```python
# =========================================
# Kaggle Ready-to-Use Notebook Template
# Auto-load and explore all uploaded CSV datasets
# =========================================

# 1️⃣ Import libraries
import os
import pandas as pd

# 2️⃣ List all files in /kaggle/input
print("All files in /kaggle/input:")
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

# 3️⃣ Automatically load all CSV files into DataFrames
dataframes = {}  # dictionary to store datasets

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename.endswith('.csv'):
            file_path = os.path.join(dirname, filename)
            try:
                # Create a safe DataFrame name from filename
                df_name = filename.replace(".csv", "").replace(" ", "_").replace("-", "_")
                dataframes[df_name] = pd.read_csv(file_path)
                print(f"✅ Loaded: {filename} as '{df_name}' DataFrame")
            except Exception as e:
                print(f"❌ Failed to load {filename}: {e}")

# 4️⃣ Explore each dataset
for name, df in dataframes.items():
    print("\n" + "="*60)
    print(f"DataFrame: {name}")
    print("="*60)
    print("Shape:", df.shape)
    print("\nColumns:", df.columns.tolist())
    
    # Show first 5 rows
    print("\nFirst 5 rows:")
    print(df.head())
    
    # Basic summary statistics for numeric columns
    print("\nSummary statistics (numeric columns):")
    print(df.describe())
    
    # Count missing values per column
    print("\nMissing values per column:")
    print(df.isnull().sum())
    
    # Optional: preview first few rows of any object/string columns
    object_cols = df.select_dtypes(include='object').columns
    if len(object_cols) > 0:
        print("\nPreview of string columns:")
        print(df[object_cols].head())
```
